# Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CSV_PATH = "./dataset/full_dataset.csv"
```

# Read file csv

```python
dataset = pd.read_csv(CSV_PATH, index_col=[0])
dataset.head()
```

# Number of datasets used and number of images in each dataset

```python
dataset['dataset'].value_counts()
```

```python
sns.barplot(x = sorted(dataset.dataset.unique()), y = dataset.groupby('dataset')['dataset'].count())
```

# Number of species

```python
len(dataset.wood_name.unique())
```

# Number of images per species

```python
pd.options.display.max_rows = 100
dataset.wood_name.value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x = dataset.wood_name.value_counts(), y = sorted(dataset.wood_name.unique()))
plt.rc('ytick', labelsize=5) 
plt.rc('xtick', labelsize=10)
plt.show()
```
